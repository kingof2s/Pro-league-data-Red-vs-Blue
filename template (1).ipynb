```python
import pandas as pd
import numpy as np
import os

import plotly.express as px
pd.options.plotting.backend = 'plotly'
```

### Cleaning and EDA

```python
fp = r"data\2014_LoL_esports_match_data_from_OraclesElixir.csv"
df = pd.read_csv(fp, index_col=0)
```

```python
import os
from tqdm.notebook import tqdm
```

```python
directory = r'data'

df = pd.DataFrame()
for filename in tqdm(os.listdir(directory)):
    new_df = pd.read_csv(os.path.join(directory, filename), index_col=0)
    df = pd.concat([df, new_df])
df
```

```python
df
```

```python
def champ_variety(group):
    top_10 = group.value_counts().head(10)
    return np.std(top_10)

(df.groupby('position')
 .agg({
     'champion': champ_variety,
 })
 .sort_values('champion')
 .rename(columns={'champion': 'Champion Variety (Std)'})
)
```

```python
df.describe()
```

```python
(df[['teamname', 'url' ,'league', 'side', 'inhibitors', 'opp_inhibitors', 'result']]
 .reset_index()
 .groupby(['gameid','teamname'])
 .agg({
        'url': 'first',
        'league': 'first',
        'side': 'first',
        'inhibitors': 'sum',
        'opp_inhibitors': 'sum',
        'result': 'first'
    })
 .sort_values("inhibitors", ascending=False)
 .assign(result=lambda x: x['result'].map({1: 'Win', 0: 'Loss'}))
 .iloc[0]
 )
```
